```python
print("üìä RESUMEN GENERAL")
for year in range(2021, 2022):
    for month in range(1, 13):
        folder_name = f"{year}_{month:02d}"
        folder_path = os.path.join(base_path, folder_name)

        if not os.path.isdir(folder_path):
            continue

        try:
            files = [f for f in os.listdir(folder_path) if f.endswith('.tif')]
            present_codes = set(f.split('_')[1] for f in files if '_' in f)

            missing = expected_set - present_codes
            unexpected = present_codes - expected_set

            print(f"üìÇ {folder_name} ‚Üí encontrados: {len(present_codes)} | faltantes: {len(missing)} | sobrantes: {len(unexpected)}")

            detalles.append({
                "folder": folder_name,
                "missing": sorted(missing, key=lambda c: expected_codes.get(c, 9999)),
                "unexpected": sorted(unexpected)
            })

        except Exception as e:
            print(f"‚ùå Error en {folder_name}: {e}")
```

```python
print("\nüßæ DETALLE POR CARPETA")
for detalle in detalles:
    print(f"\nüìÅ Carpeta: {detalle['folder']}")
    print(f"‚ùå C√≥digos FALTANTES: {len(detalle['missing'])}")
    for code in detalle['missing']:
        print(f"   FALTANTE ‚Üí {code}")
    print(f"‚ö†Ô∏è C√≥digos NO ESPERADOS: {len(detalle['unexpected'])}")
    for code in detalle['unexpected']:
        print(f"   NO ESPERADO ‚Üí {code}")
```
